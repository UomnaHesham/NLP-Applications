```python
import os, re, gc, torch, pandas as pd, nltk
from nltk.corpus import stopwords
from typing import List, Dict, Any
from datasets import Dataset
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.schema import Document
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from langchain.memory import ConversationBufferMemory
from langchain.agents import Tool, initialize_agent, AgentExecutor
from langchain.agents.agent_types import AgentType
from langchain.agents import AgentOutputParser
from langchain.schema import AgentAction, AgentFinish
from langchain.llms.base import LLM
import google.generativeai as genai


nltk.download("stopwords")
stop_words = set(stopwords.words("english"))
```

```python
class MedicalOutputParser(AgentOutputParser):
    def parse(self, text: str) -> AgentAction | AgentFinish:
        if "Final Answer:" in text:
            return AgentFinish(
                return_values={"output": text.split("Final Answer:")[-1].strip()},
                log=text,
            )
        elif "Action:" in text:
            action = text.split("Action:")[-1].split("Action Input:")[0].strip()
            action_input = text.split("Action Input:")[-1].strip()
            return AgentAction(action.strip(), action_input.strip(), text)
        else:
            return AgentFinish(
                return_values={"output": text.strip()},
                log=text,
            )


def preprocess_text(text):
    text = re.sub(r"http\S+|www\S+|https\S+", "", text, flags=re.MULTILINE)
    text = re.sub(r"<.*?>", "", text)
    text = text.lower()
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text
```

```python
df = pd.read_csv("train_data_chatbot.csv", nrows=3000)
df["short_question"] = df["short_question"].apply(preprocess_text)
df["short_answer"] = df["short_answer"].apply(preprocess_text)
df["tags"] = df["tags"].apply(lambda x: eval(x) if isinstance(x, str) else x)
df["text"] = df.apply(lambda r: f"Q: {r['short_question']}\nTags: {', '.join(r['tags'])}\nA: {r['short_answer']}", axis=1)
```

```python
documents = [Document(page_content=row["text"]) for _, row in df.iterrows()]
embedding_model = HuggingFaceEmbeddings()
vectorstore = Chroma.from_documents(documents, embedding_model, persist_directory="./chroma_store")
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
# ----------------- Gemini Text Generation -----------------
def gemini_generate(prompt: str) -> str:
    response = gemini_model.generate_content(prompt)
    return response.text.strip()

# ----------------- Gemini LLM Wrapper -----------------
class GeminiLLM(LLM):
    def _call(self, prompt: str, stop=None) -> str:
        return gemini_generate(prompt)

    @property
    def _llm_type(self) -> str:
        return "gemini"

llm = GeminiLLM()
```

```python


prompt_summarize = PromptTemplate(
    input_variables=["knowledge"],
    template="You are a medical summarizer. Summarize and structure this medical information clearly:\n\nKnowledge:\n{knowledge}\n\nSummary:"
)

prompt_answer = PromptTemplate(
    input_variables=["summary", "question"],
    template="You are a final medical expert. Using the summarized knowledge, answer the user question clearly and professionally.\n\nSummary:\n{summary}\n\nQuestion:\n{question}\n\nAnswer:"
)


def retrieve_knowledge(query: str) -> str:
    docs = retriever.get_relevant_documents(query)
    return "\n".join(doc.page_content for doc in docs)

def summarize_knowledge(knowledge: str) -> str:
    prompt = prompt_summarize.format(knowledge=knowledge)
    return gemini_generate(prompt)

def generate_final_answer(inputs: Dict[str, str]) -> str:
    prompt = prompt_answer.format(summary=inputs["summary"], question=inputs["question"])
    return gemini_generate(prompt)
```

```python

retriever_tool = Tool(
    name="RetrieverAgent",
    func=retrieve_knowledge,
    description="Retrieves relevant medical knowledge for a given question."
)

summarizer_tool = Tool(
    name="SummarizerAgent",
    func=summarize_knowledge,
    description="Summarizes retrieved medical knowledge into structured information."
)

answer_tool = Tool(
    name="AnswerGeneratorAgent",
    func=generate_final_answer,
    description="Generates final polished medical answer based on summary and user question."
)
```

```python
memory = ConversationBufferMemory(memory_key="chat_history")

# ----------------- Setup AgentExecutor -----------------
# Chain the three agents to work together
tools = [retriever_tool, summarizer_tool, answer_tool]

multi_agent_executor = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
    memory=memory,
    handle_parsing_errors=True,
    agent_kwargs={
        'output_parser': MedicalOutputParser(),
        'system_message': """You are a team of three agents working together:

1. RetrieverAgent: Find relevant medical knowledge related to the question.
2. SummarizerAgent: Summarize the retrieved knowledge clearly.
3. AnswerGeneratorAgent: Craft the final complete answer.

Follow this pipeline strictly. Communicate clearly and collaboratively. Always ensure the final answer is accurate and medically professional.

When answering, use:

Question: {the input}
Thought: what you will do
Action: {RetrieverAgent/SummarizerAgent/AnswerGeneratorAgent}
Action Input: {input}
Observation: {tool output}
Thought: what's next
Final Answer: {your final output}
"""
    },
    max_iterations=5
)
```

```python
question = "What are symptoms of hypertension in elderly?"
print(f"\nQuestion: {question}")
response = multi_agent_executor.run(question)
print(f"Answer: {response}")
```

```python

val_df = pd.read_csv("/content/validation_data_chatbot.csv", nrows=10)
val_df["short_question"] = val_df["short_question"].apply(preprocess_text)
val_df["short_answer"] = val_df["short_answer"].apply(preprocess_text)


generated_answers = []
real_answers = []

for idx, row in val_df.iterrows():
    question = row["short_question"]
    real_answer = row["short_answer"]

    try:
        generated_answer = multi_agent_executor.run(question)
    except Exception as e:
        print(f"Error for question {idx}: {e}")
        generated_answer = ""

    generated_answers.append(generated_answer)
    real_answers.append(real_answer)


from bert_score import score

P, R, F1 = score(generated_answers, real_answers, lang="en", verbose=True)


print("\n--- Validation Results ---")
print(f"Average Precision: {P.mean().item():.4f}")
print(f"Average Recall: {R.mean().item():.4f}")
print(f"Average F1 Score: {F1.mean().item():.4f}")
```
